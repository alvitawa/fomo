```python
import numpy as np
```

```python

with open("data_ablations.txt", "r") as file:
    # Iterate over each line in the file


    base, novel = None, None
    std_base_dict = {}
    std_novel_dict = {}
    context_base_dict = {}
    context_novel_dict = {}
    momentum_base_dict = {}
    momentum_novel_dict = {}
    epoch_base_dict = {}
    epoch_novel_dict = {}


    for i,data_point in enumerate(file):
        if i % 2 == 0:
            if data_point == '':
                continue
            else:
                base = float(data_point.split(",")[0].strip('[').strip(']'))
                novel = float(data_point.split(",")[1].replace(']', '').replace('[', ''))
                print(base, novel)
        else:
            if data_point == '':
                continue
            else:
                # Split the line and parse the parameters
                parts = data_point.split('_')
                momentum = float(parts[1].replace('m', ''))
                std = float(parts[2].replace('wstd', ''))
                context = float(parts[3].replace('nctxt', ''))
                max_epoch = int(parts[-1].replace('maxepoch', ''))
                
                # Update dictionaries using setdefault to avoid repetitive code
                std_base_dict.setdefault(std, []).append(base)
                std_novel_dict.setdefault(std, []).append(novel)
                context_base_dict.setdefault(context, []).append(base)
                context_novel_dict.setdefault(context, []).append(novel)
                momentum_base_dict.setdefault(momentum, []).append(base)
                momentum_novel_dict.setdefault(momentum, []).append(novel)
                epoch_base_dict.setdefault(max_epoch, []).append(base)
                epoch_novel_dict.setdefault(max_epoch, []).append(novel)
                                
                print(momentum, std, context, max_epoch)
```

```python
std_results = {}
std_base_all_values = []
std_novel_all_values = []
for key in std_base_dict:
    # print(key, sum(std_base_dict[key])/len(std_base_dict[key]), sum(std_novel_dict[key])/len(std_novel_dict[key]))
    std_base = sum(std_base_dict[key])/len(std_base_dict[key])
    std_base_all_values.append(std_base)
    std_novel = sum(std_novel_dict[key])/len(std_novel_dict[key])
    std_novel_all_values.append(std_novel)
    std_results[key] = (std_base, std_novel)
    # print(f"Std: {key} -> Average Accuracy Base: {std_base:.2f}, Average Accuracy Novel: {std_novel:.2f}")

print(np.std(std_base_all_values), np.std(std_novel_all_values))
print(std_results)
```

```python
context_results = {}
context_base_all_values = []
context_novel_all_values = []
for key in context_base_dict:
    context_base = sum(context_base_dict[key])/len(context_base_dict[key])
    context_base_all_values.append(context_base)
    context_novel = sum(context_novel_dict[key])/len(context_novel_dict[key])
    context_novel_all_values.append(context_novel)
    context_results[key] = (context_base, context_novel)
print(context_results)
print(np.std(context_base_all_values), np.std(context_novel_all_values))
```

```python
momentum_results = {}
momentum_base_all_values = []
momentum_novel_all_values = []
for key in momentum_base_dict:
    momentum_base = sum(momentum_base_dict[key])/len(momentum_base_dict[key])
    momentum_base_all_values.append(momentum_base)
    momentum_novel = sum(momentum_novel_dict[key])/len(momentum_novel_dict[key])
    momentum_novel_all_values.append(momentum_novel)
    momentum_results[key] = (momentum_base, momentum_novel)
print(momentum_results)
print(np.std(momentum_base_all_values), np.std(momentum_novel_all_values))
```

```python
epoch_results = {}
epoch_base_all_values = []
epoch_novel_all_values = []
for key in epoch_base_dict:
    epoch_base = sum(epoch_base_dict[key])/len(epoch_base_dict[key])
    epoch_base_all_values.append(epoch_base)
    epoch_novel = sum(epoch_novel_dict[key])/len(epoch_novel_dict[key])
    epoch_novel_all_values.append(epoch_novel)
    epoch_results[key] = (epoch_base, epoch_novel)

print(np.std(epoch_base_all_values), np.std(epoch_novel_all_values))
print(epoch_results)
```

```python
from tabulate import tabulate

# Assuming you have dictionaries like std_results, context_results, and momentum_results

# Convert dictionaries to lists of tuples
std_data = [(key, value) for key, value in std_results.items()]
context_data = [(key, value) for key, value in context_results.items()]
momentum_data = [(key, value) for key, value in momentum_results.items()]
epoch_data = [(key, value) for key, value in epoch_results.items()]

# Generate LaTeX tables
std_table = tabulate(std_data, headers=["Std", "Results"], tablefmt="latex")
context_table = tabulate(context_data, headers=["Context", "Results"], tablefmt="latex")
momentum_table = tabulate(momentum_data, headers=["Momentum", "Results"], tablefmt="latex")
epoch_table = tabulate(epoch_data, headers=["Epoch", "Results"], tablefmt="latex")

# Print LaTeX tables
print("STD Results Table:")
print(std_table)

print("\nContext Results Table:")
print(context_table)

print("\nMomentum Results Table:")
print(momentum_table)

print("\nEpoch Results Table:")
print(epoch_table)
```
